# Question 1: What is the difference between AI, ML, DL, and Data Science? Provide a brief explanation of each.
Hint: Compare their scope, techniques, and applications for each.

***1. Artificial Intelligence (AI)***

**Scope:** AI is the broadest field, encompassing any technique that enables computers to mimic human intelligence. It's about creating intelligent agents that perceive their environment and take actions that maximize their chance of achieving their goals.

**Techniques:** Symbolic AI (expert systems, logic programming), search algorithms, planning, natural language processing, computer vision, robotics, and learning (including ML and DL).

**Applications:** Self-driving cars, virtual assistants (Siri, Alexa), game playing (Deep Blue, AlphaGo), medical diagnosis, fraud detection, and recommendation systems.

***2. Machine Learning (ML)***

**Scope:** ML is a subset of AI that focuses on enabling systems to learn from data without being explicitly programmed. The core idea is to build models that can identify patterns and make predictions or decisions based on new, unseen data.

**Techniques:** Supervised learning (regression, classification), unsupervised learning (clustering, dimensionality reduction), reinforcement learning (Q-learning, policy gradients). Algorithms include linear regression, decision trees, support vector machines (SVMs), k-nearest neighbors (KNN), and neural networks.

**Applications:** Spam filtering, image recognition, predicting stock prices, product recommendations, and medical prognoses.

***3. Deep Learning (DL)***

**Scope:** DL is a specialized subset of Machine Learning that uses neural networks with many layers (hence 'deep') to learn complex patterns from large amounts of data. It's particularly effective with unstructured data like images, audio, and text.

**Techniques:** Convolutional Neural Networks (CNNs) for image processing, Recurrent Neural Networks (RNNs) and Long Short-Term Memory (LSTM) for sequential data (like text and time series), Generative Adversarial Networks (GANs), and Transformers.

**Applications:** Advanced image and speech recognition, natural language understanding and generation, machine translation, drug discovery, and content generation.

***4. Data Science***

**Scope:** Data Science is an interdisciplinary field that uses scientific methods, processes, algorithms, and systems to extract knowledge and insights from structured and unstructured data. It's a broader field than just ML or DL, focusing on the entire lifecycle of data.

**Techniques:** Data collection, cleaning, and preprocessing; exploratory data analysis (EDA); statistical modeling; machine learning; data visualization; and communication of findings. It often involves programming (Python, R), databases (SQL), and big data tools.

**Applications:** Business intelligence, market analysis, predicting customer behavior, optimizing business processes, scientific research, and developing data-driven products.

***In summary:***

**AI** is the big umbrella of making machines intelligent.

**ML** is a way to achieve AI by making machines learn from data.

**DL** is a specific, powerful method within ML that uses deep neural networks.

**Data Science** is an overarching discipline that uses all these techniques (and more) to extract insights and value from data for various purposes.

# Question 2: Explain overfitting and underfitting in ML. How can you detect and prevent them?**
Hint: Discuss bias-variance tradeoff, cross-validation, and regularization techniques.**

**Overfitting and Underfitting in Machine Learning**

***1. Overfitting***

**Explanation:** Overfitting occurs when a machine learning model learns the training data too well, including its noise and random fluctuations, to the extent that it performs poorly on unseen data. The model becomes overly complex and essentially 'memorizes' the training examples rather than learning the underlying patterns.

**Detection:**

- High Variance: An overfit model will show a very low error on the training set but a significantly higher error on the validation or test set.

- Poor Generalization: The model fails to generalize well to new, unseen data.

- Complex Model: Often characterized by models with too many features, overly complex algorithms, or models trained for too many epochs.

**Prevention:**

- More Training Data: Providing more diverse training data can help the model learn the true underlying patterns rather than noise.

- Feature Selection/Engineering: Reducing the number of features or creating more meaningful features can simplify the model.

- Early Stopping: For iterative learning algorithms (like neural networks), stop training when the performance on a validation set starts to degrade.

- Cross-Validation: Helps to get a more reliable estimate of the model's performance on unseen data and can reveal overfitting by comparing training and validation scores.

- Regularization Techniques: These methods add a penalty term to the loss function during training, discouraging overly complex models:

  - L1 Regularization (Lasso): Adds a penalty proportional to the absolute value of the coefficients. It can lead to sparse models, effectively performing feature selection by shrinking some coefficients to zero.

  - L2 Regularization (Ridge): Adds a penalty proportional to the square of the coefficients. It helps prevent large coefficients and makes the model more robust to minor changes in input.
  - Dropout (for Neural Networks): Randomly drops a fraction of neurons during training, forcing the network to learn more robust features.

- Ensemble Methods: Techniques like Bagging (e.g., Random Forests) can reduce variance and thus overfitting.

***2. Underfitting***

**Explanation:** Underfitting occurs when a machine learning model is too simple to capture the underlying patterns in the training data. It performs poorly on both the training data and unseen data because it hasn't learned enough from the data.

**Detection:**

- High Bias: An underfit model will show high error on both the training set and the validation/test set.

- Oversimplified Model: The model is not complex enough to represent the relationships in the data.

**Prevention:**

- More Complex Model: Use a more powerful or flexible model (e.g., switch from linear regression to a polynomial regression or a more complex neural network architecture).

- More Features: Add more relevant features to the model, or perform more sophisticated feature engineering.

- Reduce Regularization: If regularization was applied, reduce its strength or remove it.

- Increase Training Time/Epochs: For iterative algorithms, ensure the model is trained for enough iterations to converge.

- Increase Model Complexity: For neural networks, add more layers or neurons.

**Bias-Variance Tradeoff**

Overfitting and underfitting are deeply connected to the bias-variance tradeoff, a central concept in machine learning:

- Bias: Refers to the error introduced by approximating a real-world problem, which may be complex, by a simplified model. High bias leads to underfitting (the model consistently misses the mark).

- Variance: Refers to the amount that the estimate of the target function will change if different training data were used. High variance leads to overfitting (the model is too sensitive to the training data and generalizes poorly).


# Question 3:How would you handle missing values in a dataset? Explain at least three methods with examples.

Hint: Consider deletion, mean/median imputation, and predictive modeling.

***1. Deletion Methods***

**Explanation:** Deletion methods involve removing rows or columns that contain missing values. This is the simplest approach but can lead to significant loss of information if many values are missing.

- Listwise Deletion (Row Deletion): Removes entire rows (observations) that contain one or more missing values. This is suitable when the number of missing values is small and they are randomly distributed.

   - Pros: Simple to implement, maintains data integrity for complete cases.
   - Cons: Can lead to a substantial reduction in dataset size, potentially biasing the analysis if missingness is not completely random.

- Pairwise Deletion: Uses all available data for each specific analysis. For example, when calculating correlations, it uses all pairs of data points where both values are present.

  - Pros: Utilizes more data than listwise deletion.
  - Cons: Can lead to different sample sizes for different analyses, complicating comparisons.

```python
#Example (Python with Pandas): Let's say we have a DataFrame df with some missing values.

import pandas as pd
import numpy as np

data = {'A': [1, 2, 3, 4, np.nan],
        'B': [5, np.nan, 7, 8, 9],
        'C': [10, 11, 12, np.nan, 14]}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# Listwise Deletion
df_deleted_rows = df.dropna()
print("\nDataFrame after row deletion:")
print(df_deleted_rows)

# Column Deletion (if a column has too many missing values, e.g., >70%)
df_deleted_cols = df.dropna(axis=1, thresh=len(df)*0.7) # Keep columns with at least 70% non-missing values
print("\nDataFrame after column deletion (example: keeping cols with >70% data):")
print(df_deleted_cols)
```

***2. Imputation Methods (Mean/Median/Mode)***

**Explanation:** Imputation involves filling in missing values with estimated values. Mean, median, or mode imputation are simple statistical approaches.

- Mean Imputation: Replaces missing values with the mean of the observed values for that feature. Suitable for numerical data when the data is symmetrically distributed and outliers are not a major concern.
- Median Imputation: Replaces missing values with the median of the observed values for that feature. More robust to outliers than mean imputation and suitable for numerical data, especially skewed distributions.
- Mode Imputation: Replaces missing values with the mode (most frequent value) of the observed values for that feature. Best suited for categorical or discrete numerical data.

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

data = {'Numerical_Feature': [10, 20, np.nan, 40, 50, 100],
        'Categorical_Feature': ['A', 'B', 'A', np.nan, 'C', 'B']}
df_impute = pd.DataFrame(data)
print("Original DataFrame for Imputation:")
print(df_impute)

# Mean Imputation for Numerical_Feature
mean_imputer = SimpleImputer(strategy='mean')
df_impute['Numerical_Feature_Mean'] = mean_imputer.fit_transform(df_impute[['Numerical_Feature']])

# Median Imputation for Numerical_Feature
median_imputer = SimpleImputer(strategy='median')
df_impute['Numerical_Feature_Median'] = median_imputer.fit_transform(df_impute[['Numerical_Feature']])

# Mode Imputation for Categorical_Feature
mode_imputer = SimpleImputer(strategy='most_frequent')
df_impute['Categorical_Feature_Mode'] = mode_imputer.fit_transform(df_impute[['Categorical_Feature']]).ravel() # Added .ravel() to flatten the array

print("\nDataFrame after Mean, Median, and Mode Imputation:")
print(df_impute)
```

***3. Predictive Modeling Imputation***

**Explanation:** This advanced technique uses a machine learning model to predict the missing values based on other features in the dataset. It treats the feature with missing values as the target variable and other features as predictors.

- Methods: K-Nearest Neighbors (KNN) Imputation, Regression Imputation (e.g., using Linear Regression, Ridge, Lasso, or even more complex models like Random Forest).
  - Pros: Can produce more accurate imputations by capturing complex relationships between features, potentially leading to less biased results.
  - Cons: More computationally intensive, can be complex to implement, and can amplify errors if the predictive model itself is inaccurate.


```python
#Example (Python with Scikit-learn's KNNImputer):

import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer

data_predictive = {
    'Feature_1': [1, 2, np.nan, 4, 5],
    'Feature_2': [10, np.nan, 30, 40, 50],
    'Feature_3': [100, 200, 300, np.nan, 500]
}
df_predictive = pd.DataFrame(data_predictive)
print("Original DataFrame for Predictive Imputation:")
print(df_predictive)

# KNN Imputation (e.g., using 2 nearest neighbors)
knn_imputer = KNNImputer(n_neighbors=2)
df_imputed_knn = pd.DataFrame(knn_imputer.fit_transform(df_predictive), columns=df_predictive.columns)

print("\nDataFrame after KNN Imputation:")
print(df_imputed_knn)
```

# Question 5: Why is feature scaling important in ML? Compare Min-Max scaling and Standardization.

Hint: Explain impact on distance-based algorithms (e.g., KNN, SVM) and gradient descent.

Many machine learning algorithms perform better or converge faster when features are on a similar scale. Here are the main reasons:

1. Distance-Based Algorithms (e.g., K-Nearest Neighbors (KNN), Support Vector Machines (SVM), K-Means Clustering):

   - These algorithms calculate distances between data points to find similarities or make classifications. If features have vastly different ranges, the feature with the larger range will dominate the distance calculation, effectively overshadowing the influence of features with smaller ranges. This can lead to biased results and poor performance. For example, if one feature ranges from 1 to 1000 and another from 0 to 1, the feature with the larger range will have a disproportionately large effect on the distance metric.

2. Gradient Descent-Based Algorithms (e.g., Linear Regression, Logistic Regression, Neural Networks):

   - Algorithms that use gradient descent (or its variants) to optimize their objective function converge much faster when features are scaled. If features have different scales, the cost function will have an elongated, narrow shape (like an ellipse), causing the gradient descent algorithm to take many small, zig-zag steps towards the minimum. When features are scaled, the cost function becomes more spherical, allowing the algorithm to take more direct and larger steps, leading to faster convergence.

3. Regularization Techniques (e.g., Lasso and Ridge Regression):

   - Regularization methods penalize large coefficients to prevent overfitting. If features are not scaled, features with larger magnitudes might inadvertently receive larger penalties, even if they are not inherently more important than features with smaller magnitudes.

4. Principal Component Analysis (PCA):

   - PCA aims to find directions of maximum variance in the data. If features are not scaled, features with higher variance (often due to larger scales) will contribute disproportionately more to the principal components, potentially leading to misleading results.

# Comparison of Min-Max Scaling and Standardization
These are two of the most common feature scaling techniques:

 1. Min-Max Scaling (Normalization)

Formula: X_scaled = (X - X_min) / (X_max - X_min)

Explanation: This method scales and translates each feature individually so that it falls within a specific range, typically between 0 and 1 (or -1 and 1 if specified). It rescales the data so that the minimum value of a feature is transformed into the new minimum (e.g., 0) and the maximum value into the new maximum (e.g., 1).

Characteristics:
Range: Transforms features to a fixed range (e.g., 0 to 1).

Outliers: Highly sensitive to outliers. A single outlier can significantly compress the majority of the data into a small range, making it less effective.
Distribution: Does not change the distribution of the data (e.g., if it was skewed, it remains skewed).

When to Use:
When the algorithm specifically requires inputs in a fixed range (e.g., neural networks with sigmoid activation functions, image processing pixel intensities).
When the data is not known to contain significant outliers.

2. Standardization (Z-score Normalization)

Formula: X_scaled = (X - μ) / σ, where μ is the mean and σ is the standard deviation of the feature.

Explanation: This method scales features such that they have a mean of 0 and a standard deviation of 1. It centers the data around the mean and scales it by its standard deviation.

Characteristics:
Range: Does not bound values to a specific range. Outliers will still exist but will be scaled relative to the mean and standard deviation.

Outliers: Less affected by outliers than Min-Max scaling because it uses the mean and standard deviation, which are somewhat influenced by outliers, but not to the extent of fixed min/max values.

Distribution: Transforms the data to have a standard normal distribution if the original data was normally distributed. Even for non-normal distributions, it centers and scales the data.

When to Use:
- When the algorithm assumes a normal distribution or works better with zero-mean, unit-variance data (e.g., Linear Regression, Logistic Regression, SVMs with RBF kernel, PCA).
- When the data contains outliers, as it handles them more gracefully than Min-Max scaling.
- When you don't need data to be bounded to a specific range.

**Label Encoding**

Explanation: Label Encoding assigns a unique integer to each category in a categorical variable. For example, if a feature 'Color' has categories 'Red', 'Green', 'Blue', it might be encoded as 0, 1, 2 respectively.

Pros:

- Simplicity: Easy to implement and computationally inexpensive.
- Memory Efficiency: Uses less memory than One-Hot Encoding, especially for high-cardinality categorical variables.

Cons:

- Introduces Artificial Order: The primary drawback is that it assigns an arbitrary numerical order to categories. This can mislead models into assuming an ordinal relationship where none exists. For example, if 'Red' is 0, 'Green' is 1, and 'Blue' is 2, a model might incorrectly infer that 'Blue' is 'greater' or 'more important' than 'Red'. This can negatively impact model performance, particularly for algorithms that interpret numerical differences (e.g., linear models, distance-based algorithms like KNN).

**When to Prefer Label Encoding:**

- Ordinal Categorical Variables: When the categorical variable inherently has an ordered relationship (e.g., 'Small' < 'Medium' < 'Large' can be encoded as 0, 1, 2). In such cases, the numerical representation naturally reflects the order.
- Tree-based Algorithms: Algorithms like Decision Trees, Random Forests, and Gradient Boosting Machines (e.g., XGBoost, LightGBM) are often less sensitive to the artificial order introduced by Label Encoding because they can split on individual categories without assuming a continuous numerical relationship. Some tree-based models can even handle categorical features directly, making encoding less critical.
- High Cardinality with Tree Models (or specific scenarios): If you have a very large number of unique categories and are using a tree-based model, Label Encoding might be preferred over One-Hot Encoding to avoid creating an extremely wide dataset, which could lead to increased memory usage and potentially slower training.

# Question 7: Google Play Store Dataset
# a). Analyze the relationship between app categories and ratings. Which categories have the highest/lowest average ratings, and what could be the possible reasons?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/googleplaystore.csv"
df = pd.read_csv(url)

# --- Data Cleaning ---
# Remove rows with missing or invalid ratings
df = df[pd.to_numeric(df['Rating'], errors='coerce').notnull()]
df['Rating'] = df['Rating'].astype(float)

# Remove missing categories
df = df[df['Category'].notnull()]

# --- Group by Category ---
category_ratings = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)

print("Top 5 Categories by Average Rating:")
print(category_ratings.head(), "\n")

print("Bottom 5 Categories by Average Rating:")
print(category_ratings.tail(), "\n")

# --- Visualization ---
plt.figure(figsize=(12, 7))
category_ratings.plot(kind='bar')
plt.title("Average Rating by App Category")
plt.xlabel("Category")
plt.ylabel("Average Rating")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# Question 8: Titanic Dataset

# a) Compare the survival rates based on passenger class (Pclass). Which class had the highest survival rate, and why do you think that happened?

# b) Analyze how age (Age) affected survival. Group passengers into children (Age < 18) and adults (Age ≥ 18). Did children have a better chance of survival?

```python
import pandas as pd

# Load the Titanic dataset from a public URL
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Drop rows with missing values in key columns
df = df.dropna(subset=['Survived', 'Pclass', 'Age'])

# Ensure correct data types
df['Survived'] = df['Survived'].astype(int)
df['Pclass'] = df['Pclass'].astype(int)

# -----------------------------
# Part A: Survival by Pclass
# -----------------------------
survival_by_class = df.groupby('Pclass')['Survived'].mean() * 100

print("=== Survival Rate by Passenger Class ===")
for cls, rate in survival_by_class.items():
    print(f"Class {cls}: {rate:.2f}%")

# -----------------------------
# Part B: Children vs Adults
# -----------------------------
df['AgeGroup'] = df['Age'].apply(lambda x: 'Child' if x < 18 else 'Adult')

survival_by_agegroup = df.groupby('AgeGroup')['Survived'].mean() * 100

print("\n=== Survival Rate by Age Group ===")
for group, rate in survival_by_agegroup.items():
    print(f"{group}: {rate:.2f}%")

# -----------------------------
# Optional: Display result tables
# -----------------------------
print("\nSurvival by Class Table:")
print(survival_by_class)

print("\nSurvival by Age Group Table:")
print(survival_by_agegroup)
```

# Question 9: Flight Price Prediction Dataset

# a) How do flight prices vary with the days left until departure? Identify any exponential price surges and recommend the best booking window.

# b)Compare prices across airlines for the same route (e.g., Delhi-Mumbai). Which airlines are consistently cheaper/premium, and why?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/refs/heads/main/flight_price.csv"
df = pd.read_csv(url)

# Show columns and first few rows to inspect structure
print("Columns:", df.columns.tolist())
print(df.head())

# ASSUMPTIONS:
#   - There is a column like 'Days_Until_Departure' or 'Days_left' or similar
#   - There is a column 'Price' (fare)
#   - There is a column 'Airline'
#   - There is route columns, e.g. 'Source' and 'Destination' OR a combined 'Route'
# Adapt the code below according to actual column names.

# --- Part (a): Price vs days until departure ---

# Clean -- drop rows with missing values in critical columns
# Corrected column names: 'price' and 'days_left'
df2 = df.dropna(subset=['price', 'days_left'])

# Convert to numeric if needed
df2['price'] = pd.to_numeric(df2['price'], errors='coerce')
df2['days_left'] = pd.to_numeric(df2['days_left'], errors='coerce')
df2 = df2.dropna(subset=['price', 'days_left'])

# Aggregate — compute average price for each days-left value, or bins
# Corrected column names: 'days_left' and 'price'
price_by_days = df2.groupby('days_left')['price'].mean().reset_index()

# Plot
plt.figure(figsize=(10,6))
sns.lineplot(data=price_by_days, x='days_left', y='price')
plt.title("Average Flight Price vs Days Until Departure")
plt.xlabel("Days Until Departure")
plt.ylabel("Average Price")
plt.grid(True)
plt.show()

# Optionally, use a rolling or smoothing to detect surges
price_by_days['Price_smooth'] = price_by_days['price'].rolling(window=5, min_periods=1).mean()
plt.figure(figsize=(10,6))
sns.lineplot(data=price_by_days, x='days_left', y='Price_smooth')
plt.title("Smoothed: Flight Price vs Days Until Departure")
plt.xlabel("Days Until Departure")
plt.ylabel("Smoothed Avg Price")
plt.grid(True)
plt.show()


# --- Part (b): Compare airlines on same route ---

# Let's pick a route, e.g. "Delhi" to "Mumbai"
src = "Delhi"
dst = "Mumbai"

# Filter for that route — adjust columns accordingly
# Corrected column names: 'source_city', 'destination_city', 'airline', 'price'
df_route = df[(df['source_city'] == src) & (df['destination_city'] == dst)].dropna(subset=['airline', 'price'])

# Compute average price per airline on this route
# Corrected column name: 'airline', 'price'
price_by_airline = df_route.groupby('airline')['price'].agg(['mean', 'median', 'count']).sort_values('mean')

print("\n--- Price comparison for route", src, "->", dst, "---")
print(price_by_airline)

# Plot boxplot to show price distribution per airline
# Corrected column name: 'airline', 'price'
plt.figure(figsize=(12,6))
sns.boxplot(data=df_route, x='airline', y='price')
plt.title(f"Price distribution by Airline for {src} → {dst}")
plt.xlabel("Airline")
plt.ylabel("Price")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/e0da5fc18ebf4e3dc90950b932d6bded61e0ee49/hr_analytics.csv"
df = pd.read_csv(url)

# Show columns & sample
print("Columns:", df.columns.tolist())
print(df.head())

# --- Basic Cleaning: drop rows with missing key values if any ---
df = df.dropna(subset=['left', 'satisfaction_level', 'number_project', 'average_montly_hours', 'time_spend_company', 'salary'])

# Convert types if needed
df['left'] = df['left'].astype(int)
df['number_project'] = df['number_project'].astype(int)

# --- Part (a): What factors correlate strongly with attrition (left) ---

# 1. Correlation matrix (only numeric features + target)
numeric_cols = ['satisfaction_level', 'last_evaluation', 'number_project',
                'average_montly_hours', 'time_spend_company']
corr = df[numeric_cols + ['left']].corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation matrix: Numeric features vs Attrition (left)")
plt.show()

# 2. Boxplots / distribution plots for important factors by left/not-left
# Separate numerical and categorical factors for appropriate plotting
numerical_factors_for_boxplot = ['satisfaction_level', 'number_project', 'average_montly_hours', 'time_spend_company']
categorical_factors_for_boxplot = ['salary'] # 'salary' is categorical

for f in numerical_factors_for_boxplot:
    plt.figure(figsize=(6,4))
    sns.boxplot(data=df, x='left', y=f)
    plt.title(f"Distribution of {f} by Attrition (left)")
    plt.xlabel("Left (1 = left, 0 = stayed)")
    plt.ylabel(f)
    plt.show()

# For categorical factors like 'salary', use countplots or similar
for f in categorical_factors_for_boxplot:
    plt.figure(figsize=(7,5))
    sns.countplot(data=df, x=f, hue='left', palette='viridis')
    plt.title(f"Distribution of {f} by Attrition (left)")
    plt.xlabel(f)
    plt.ylabel("Count")
    plt.show()

# 3. Mean values for left vs stayed (only for numerical factors)
mean_by_left = df.groupby('left')[numerical_factors_for_boxplot].mean()
print("\nMean values of numerical factors for stayed vs left employees:")
print(mean_by_left)

# For 'salary', we can look at the distribution counts or percentages
salary_by_left = df.groupby(['salary', 'left']).size().unstack(fill_value=0)
salary_by_left_percent = salary_by_left.apply(lambda x: x / x.sum(), axis=0)
print("\nSalary distribution for stayed vs left employees (counts):")
print(salary_by_left)
print("\nSalary distribution for stayed vs left employees (percentages):")
print(salary_by_left_percent)

# --- Part (b): Are employees with more projects more likely to leave? ---

# Let's compute attrition rate per number of projects
proj_attrition = df.groupby('number_project')['left'].mean() * 100
proj_counts = df['number_project'].value_counts().sort_index()

plt.figure(figsize=(8,5))
sns.lineplot(x=proj_attrition.index, y=proj_attrition.values, marker='o')
plt.title("Attrition rate vs Number of Projects")
plt.xlabel("Number of projects")
plt.ylabel("Attrition rate (%)")
plt.grid(True)
plt.show()

print("\nAttrition rate by number of projects:")
print(pd.concat([proj_counts, proj_attrition.rename('attrition_rate_percent')], axis=1).sort_index())
```
